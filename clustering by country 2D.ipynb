```python
# for all countries
test_df = pd.DataFrame(df,columns=['O18','H2','Country'])
groups = test_df.groupby("Country")
```

https://plotly.com/python/shapes/

```python
# plot (plotly)
country_list = ['CN','US','ZM']

# Create figure
fig = go.Figure()
index = 0
for name, group in groups:
    if name in country_list:
        # reset color index
        if index > 7: index = 0

        fig.add_trace(go.Scatter(x=group["O18"],y=group["H2"], mode="markers", marker=dict(color=px.colors.sequential.Rainbow_r[index])))
        
        # Add shapes
        fig.add_shape(type="circle",
            xref="x", yref="y",
            x0=min(group["O18"]), y0=min(group["H2"]),
            x1=max(group["O18"]), y1=max(group["H2"]),
            opacity=0.2,
            fillcolor=px.colors.sequential.Rainbow_r[index]
        )

        fig.add_trace(go.Surface)

        index += 2

# Hide legend
fig.update_layout(showlegend=False)


fig.show()
```
